```python
import pandas as pd
import numpy as np

from IPython.display import display, HTML
pd.options.mode.chained_assignment = None  # default='warn'

import matplotlib.pyplot as plt
plt.close('all')
```

```python
# Import CSV
data = pd.read_csv("original.csv", sep=';', decimal=",", encoding = "ISO-8859-1", error_bad_lines=False)
```

```python
# Display database
display(data)
```

```python
# Select only women
homem = data[(data['Sexo'] == 'M')]
mulher = data[(data['Sexo'] == 'F')]

total_homem = len(homem.index)
total_mulher = len(mulher.index)

h = homem
m = mulher

filtered_homem = homem[
    (h['Regiao_Abertura'] == 'Mordor') |
    (h['Regiao_Abertura'] == 'Gotham City') |
    (h['Regiao_Abertura'] == 'Mushroom Kingdom') |
    (h['Regiao_Abertura'] == 'Hogsmeade') |
    (h['Regiao_Abertura'] == 'Zion') |
    (h['Regiao_Fechamento'] == 'Mordor') |
    (h['Regiao_Fechamento'] == 'Gotham City') |
    (h['Regiao_Fechamento'] == 'Mushroom Kingdom') |
    (h['Regiao_Fechamento'] == 'Hogsmeade') |
    (h['Regiao_Fechamento'] == 'Zion')
]

filtered_mulher = mulher[
    (m['Regiao_Abertura'] == 'Mordor') |
    (m['Regiao_Abertura'] == 'Gotham City') |
    (m['Regiao_Abertura'] == 'Mushroom Kingdom') |
    (m['Regiao_Abertura'] == 'Hogsmeade') |
    (m['Regiao_Abertura'] == 'Zion') |
    (m['Regiao_Fechamento'] == 'Mordor') |
    (m['Regiao_Fechamento'] == 'Gotham City') |
    (m['Regiao_Fechamento'] == 'Mushroom Kingdom') |
    (m['Regiao_Fechamento'] == 'Hogsmeade') |
    (m['Regiao_Fechamento'] == 'Zion')
]

total_homem_trabalho = len(filtered_homem.index)
total_mulher_trabalho = len(filtered_mulher.index)

print("Total de homens:", total_homem)
print("Total de mulheres:", total_mulher)

print("===")

print("Homens em trabalho:", total_homem_trabalho)
print("Mulheres em trabalho:", total_mulher_trabalho)

print("===")

print("Homens em trabalho:", (total_homem_trabalho/total_homem))
print("Mulheres em trabalho:", (total_mulher_trabalho/total_mulher))
```

```python
# Store women age

item = 'Data_Nascimento'

now = pd.Timestamp('now')
women[item] = pd.to_datetime(women[item])
women[item] = women[item].where(women[item] < now, women[item] -  np.timedelta64(100, 'Y'))
women['Idade'] = (now - women[item]).astype('<m8[Y]')

display(women)
```

```python
# Get mean invested value grouped by age
# pd.to_numeric(women['Valor_Diarias'])
mean_value_by_age = women.groupby('Idade', as_index=False)['Valor_Diarias'].mean()

plt.bar(mean_value_by_age['Idade'], mean_value_by_age['Valor_Diarias'])
```

```python
# Filter women by age

women_filtered = women[(women['Idade'] > 22) & (women['Idade'] < 40)]
```

```python
# Get mean invested value grouped by age

filtered_mean_value_by_age = women_filtered.groupby('Idade', as_index=False)['Valor_Diarias'].mean()

plt.bar(filtered_mean_value_by_age['Idade'], filtered_mean_value_by_age['Valor_Diarias'])
```
