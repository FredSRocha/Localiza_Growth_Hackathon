```python
import pandas as pd
import numpy as np

from IPython.display import display, HTML
pd.options.mode.chained_assignment = None  # default='warn'

import matplotlib.pyplot as plt
plt.close('all')
```

```python
# Import CSV
data = pd.read_csv("original.csv", sep=';', decimal=",", encoding = "ISO-8859-1", error_bad_lines=False)
```

```python
# Display database
display(data)
```

```python
# Select only women
women = data[(data['Sexo'] == 'M')]
display(women)
```

```python
# Store women age

item = 'Data_Nascimento'

now = pd.Timestamp('now')
women[item] = pd.to_datetime(women[item])
women[item] = women[item].where(women[item] < now, women[item] -  np.timedelta64(100, 'Y'))
women['Idade'] = (now - women[item]).astype('<m8[Y]')

display(women)
```

```python
# Get mean invested value grouped by age
# pd.to_numeric(women['Valor_Diarias'])
mean_value_by_age = women.groupby('Idade', as_index=False)['Valor_Diarias'].mean()

plt.bar(mean_value_by_age['Idade'], mean_value_by_age['Valor_Diarias'])
```

```python
# Filter women by age

women_filtered = women[(women['Idade'] > 22) & (women['Idade'] < 40)]
```

```python
# Get mean invested value grouped by age

filtered_mean_value_by_age = women_filtered.groupby('Idade', as_index=False)['Valor_Diarias'].mean()

plt.bar(filtered_mean_value_by_age['Idade'], filtered_mean_value_by_age['Valor_Diarias'])
```
